```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, average_precision_score
)

from sklearn.ensemble import RandomForestClassifier
```

```python
CSV_PATH = r"C:\Users\antho\Desktop\CUD\Semester 6\Poster research\VeReMi_Extension Dataset for Misbehaviors in VANETs\VeReMi_Extension Dataset for Misbehaviors in VANETs\mixalldata_clean\mixalldata_clean.csv"  # <-- change this
df = pd.read_csv(CSV_PATH)

print("Loaded:", df.shape)
print("Classes:", sorted(df["class"].unique()))
print(df["class"].value_counts().head(10))
print(df.columns)
```

```python
# Binary label: 0 = benign, 1 = attack
df["y_binary"] = (df["class"] != 0).astype(int)

# -----------------------------
# Observed = base + noise
# -----------------------------

# Speed
df["spd_obs_x"] = df["spdx"] + df["spdx_n"]
df["spd_obs_y"] = df["spdy"] + df["spdy_n"]

# Acceleration
df["acl_obs_x"] = df["aclx"] + df["aclx_n"]
df["acl_obs_y"] = df["acly"] + df["acly_n"]

# Heading
df["hed_obs_x"] = df["hedx"] + df["hedx_n"]
df["hed_obs_y"] = df["hedy"] + df["hedy_n"]

# Position  ✅ ADDED
df["pos_obs_x"] = df["posx"] + df["posx_n"]
df["pos_obs_y"] = df["posy"] + df["posy_n"]

# -----------------------------
# Features (now including position)
# -----------------------------

FEATURES = [
    "pos_obs_x", "pos_obs_y",     # ← position added
    "spd_obs_x", "spd_obs_y",
    "acl_obs_x", "acl_obs_y",
    "hed_obs_x", "hed_obs_y"
]

X = df[FEATURES].copy()
y = df["y_binary"].copy()

print("X shape:", X.shape)
print("y counts:\n", y.value_counts())
print("Any NaNs in X?", X.isna().any().any())
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

RANDOM_STATE = 42
KNOWN_ATTACKS   = [16, 13]
UNKNOWN_ATTACKS = [18, 19]

df = df.copy()
df["y_binary"] = (df["class"] != 0).astype(int)

# indices
benign_idx         = df.index[df["class"] == 0]
known_attack_idx   = df.index[df["class"].isin(KNOWN_ATTACKS)]
unknown_attack_idx = df.index[df["class"].isin(UNKNOWN_ATTACKS)]

# ✅ split benign into two disjoint pools
benign_A_idx, benign_B_idx = train_test_split(
    benign_idx,
    test_size=0.30,          # 30% benign reserved for UNKNOWN
    random_state=RANDOM_STATE,
    shuffle=True
)

# KNOWN dataset for CV/training
known_idx = df.index[df.index.isin(benign_A_idx) | df.index.isin(known_attack_idx)]
X = df.loc[known_idx, FEATURES].copy()
y = df.loc[known_idx, "y_binary"].copy()

# UNKNOWN holdout (zero-day) with different benign
unknown_idx = df.index[df.index.isin(benign_B_idx) | df.index.isin(unknown_attack_idx)]
X_unknown = df.loc[unknown_idx, FEATURES].copy()
y_unknown = df.loc[unknown_idx, "y_binary"].copy()

print("Known y counts:\n", y.value_counts())
print("\nUnknown y counts:\n", y_unknown.value_counts())
print("\n✅ benign overlap (must be 0):", len(set(benign_A_idx) & set(benign_B_idx)))
print("\nUnknown class counts:\n", df.loc[unknown_idx, "class"].value_counts())
```

```python
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, average_precision_score
)

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("rf", RandomForestClassifier(
        n_estimators=300,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        class_weight="balanced"
    ))
])

cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=RANDOM_STATE)

scoring = {
    "accuracy": "accuracy",
    "precision": "precision",
    "recall": "recall",
    "f1": "f1",
    "roc_auc": "roc_auc",
    "pr_auc": "average_precision"
}

# ===== known metrics via CV (this is what you report for known) =====
cv_res = cross_validate(pipe, X, y, cv=cv, scoring=scoring)

print("===== 4-Fold Cross Validation (KNOWN) =====")
for m in scoring:
    print(f"{m}: {np.mean(cv_res['test_'+m]):.4f} ± {np.std(cv_res['test_'+m]):.4f}")

# Train final model on ALL known data (after CV)
pipe.fit(X, y)

# ===== unknown metrics (zero-day) =====
proba_u = pipe.predict_proba(X_unknown)[:, 1]
pred_u  = (proba_u >= 0.5).astype(int)

print("\n===== UNKNOWN (Zero-Day Test) =====")
print("accuracy :", round(accuracy_score(y_unknown, pred_u), 4))
print("precision:", round(precision_score(y_unknown, pred_u, zero_division=0), 4))
print("recall   :", round(recall_score(y_unknown, pred_u, zero_division=0), 4))
print("f1       :", round(f1_score(y_unknown, pred_u, zero_division=0), 4))
print("roc_auc  :", round(roc_auc_score(y_unknown, proba_u), 4))
print("pr_auc   :", round(average_precision_score(y_unknown, proba_u), 4))
```
